```python
def flatten(lst):
    return [item for sublist in lst if type(sublist)==list for item in sublist  ]
    
```

```python
import pandas as pd 
import os
import numpy as np
```

```python
all_df = pd.read_csv('data/all_admission_fulladm_all_stats.csv', index_col=0)
```

```python
# len(all_df['mean_heightcm'].dropna())

subjects_df =  pd.read_csv('../../LHC_mimic/mimic3_1.4/raw_data/PATIENTS.csv')[['SUBJECT_ID','DOB','GENDER']]
adm_df = pd.read_csv('../../LHC_mimic/mimic3_1.4/raw_data/ADMISSIONS.csv').merge(subjects_df, on='SUBJECT_ID', how='left')

adm_df['ADMITTIME'] = pd.to_datetime(adm_df['ADMITTIME']).dt.date
adm_df['DOB'] = pd.to_datetime(adm_df['DOB'] ).dt.date

adm_df['AGE'] = adm_df.apply(lambda row: (row['ADMITTIME']-row['DOB']).days / 365.25,axis=1)

all_df = all_df.merge(adm_df[['HADM_ID','GENDER','AGE']], on='HADM_ID', how='outer')

# stays_df = pd.read_csv('../../LHC_mimic/mimic3_1.4/raw_data/ICUSTAYS.csv')
# all_df =  all_df.merge(stays_df[['HADM_ID','ICUSTAY_ID']], how='left', on='HADM_ID')

# all_df[all_df['HADM_ID']==104557][[col for col in all_df.columns if 'weight' in col or'height' in col]]
```

```python
all_df
```

```python
all_df['MED_ACE'] = all_df['MED_ACE'].fillna(False).apply(lambda x: 1.0 if x else 0.0)
all_df['MED_BETA'] = all_df['MED_BETA'].fillna(False).apply(lambda x: 1.0 if x else 0.0)
all_df['MED_INS'] = all_df['MED_INS'].fillna(False).apply(lambda x: 1.0 if x else 0.0)
all_df['diabetesdx'] = all_df['diabetesdx'].fillna(0.0)
```

```python
rf_identifiers = {
    'HADM_ID':lambda row: row['HADM_ID'],
    'HYPERTENSION':lambda row: row['max_sysbp']>140 or row['max_diasbp']>90,
    'HYPERLIPIDEMIA': lambda row: row['max_ldl']>100 or row['max_cholesterol']>240,
    'MEDICATION': lambda row: row['MED_ACE']==1 or row['MED_BETA']==1 or row['MED_INS']==1,
    'DIABETES': lambda row: row['diabetesdx']==1,
    'OBESE': lambda row: row['max_weightkg']/(row['max_heightcm']/100)**2 > 30
}
```

```python
rfs_df =  pd.DataFrame({rf:all_df.apply(fn,axis=1) for rf,fn in rf_identifiers.items()})
    
```

```python
rfs_df['rf_string'] = rfs_df.apply(lambda row:' '.join([ idx  for idx in row.index if row[idx]==True]), axis=1)
```

```python
rfs_df['rf_string'] = rfs_df['rf_string'].fillna('')
```

```python
rfs_df
```

```python
train_df = pd.read_csv('data/CAD_train_halfsplit.csv', index_col=0)
train_df['HADM_ID'] = train_df['HADM_ID'].apply(lambda x: int(x))
train_df = train_df.merge(rfs_df, on='HADM_ID',how='left')
train_df['rf_string'] = train_df['rf_string'].fillna('')
train_df['TEXT']= train_df.apply(lambda row: row['rf_string'] +' '+ row['TEXT'],axis=1)
# train_df.to_csv('data/CAD_train_halfsplit_leakedconcepts.csv')
```

```python
val_df = pd.read_csv('data/CAD_val_halfsplit.csv', index_col=0)
val_df['HADM_ID'] = val_df['HADM_ID'].apply(lambda x: int(x))
val_df = val_df.merge(rfs_df, on='HADM_ID',how='left')
val_df['rf_string'] = val_df['rf_string'].fillna('')
val_df['TEXT']= val_df.apply(lambda row: row['rf_string'] +' '+ row['TEXT'],axis=1)
# val_df.to_csv('data/CAD_val_halfsplit_leakedconcepts.csv')
```

```python
test_df = pd.read_csv('data/CAD_test_halfsplit.csv', index_col=0)
test_df['HADM_ID'] = test_df['HADM_ID'].apply(lambda x: int(x))
test_df = test_df.merge(rfs_df, on='HADM_ID',how='left')
test_df['rf_string'] = test_df['rf_string'].fillna('')
test_df['TEXT']= test_df.apply(lambda row: row['rf_string'] +' '+ row['TEXT'],axis=1)
# test_df.to_csv('data/CAD_test_halfsplit_leakedconcepts.csv')
```

```python
val_and_test = pd.concat([test_df,val_df],ignore_index=True).fillna(False)
```

```python
for concept in ['HYPERTENSION', 'HYPERLIPIDEMIA','MEDICATION','DIABETES','OBESE']:
    print(concept)
    val_and_test[val_and_test[concept]].to_csv('concept_listfiles/leakedconcepts_'+concept)
    for i in range(50):
        val_and_test[val_and_test[concept].apply(lambda x: not x)].sample(500).to_csv('concept_listfiles/negative_leakedconcepts_'+concept+str(i))
```

```python
all_df['GENDER']=all_df['GENDER'].map({'M':1,'F':0})
```

```python
all_df_normalized = all_df.copy()
for col in all_df.columns:
    if col in ['HADM_ID']:
        continue
    if 'count' in col:
        all_df_normalized[col]=all_df_normalized[col].fillna(0.0)
    mn = all_df_normalized[col].min()
    mx = all_df_normalized[col].max()
    all_df_normalized[col] = all_df_normalized[col].fillna(all_df_normalized[col].median())
    all_df_normalized[col] = all_df_normalized[col].apply(lambda x: x-mn)
    all_df_normalized[col] = all_df_normalized[col].apply(lambda x: x/mx)
    
```

```python
all_df_normalized
```

```python
val_df = pd.read_csv('data/CAD_val_halfsplit.csv', index_col=0)
val_df['HADM_ID'] = val_df['HADM_ID'].apply(lambda x: int(x))
val_df = val_df.merge(all_df_normalized, on='HADM_ID', how='left')
val_df['diabetesdx'] = val_df['diabetesdx'].fillna(0.0)
val_df['MED_ACE'] = val_df['MED_ACE'].fillna(0.0)
val_df['MED_INS'] = val_df['MED_INS'].fillna(0.0)
val_df['MED_BETA'] = val_df['MED_BETA'].fillna(0.0)
for col in all_df_normalized.columns:
    if col in ['HADM_ID']:
        continue
    if 'count' in col:
        val_df[col].fillna(0.0)
    val_df[col] = val_df[col].fillna(all_df_normalized[col].median())
val_df.to_csv('data/CAD_val_halfsplit_plus_structured.csv')
```

```python
val_df
```

```python
train_df = pd.read_csv('data/CAD_train_halfsplit.csv', index_col=0)
train_df['HADM_ID'] = train_df['HADM_ID'].apply(lambda x: int(x))
train_df = train_df.merge(all_df_normalized, on='HADM_ID', how='left')
train_df['diabetesdx'] = train_df['diabetesdx'].fillna(0.0)
train_df['MED_ACE'] = train_df['MED_ACE'].fillna(0.0)
train_df['MED_INS'] = train_df['MED_INS'].fillna(0.0)
train_df['MED_BETA'] = train_df['MED_BETA'].fillna(0.0)
for col in all_df_normalized.columns:
    if col in ['HADM_ID']:
        continue
    if 'count' in col:
        train_df[col].fillna(0.0)
    train_df[col] = train_df[col].fillna(all_df_normalized[col].median())
train_df.to_csv('data/CAD_train_halfsplit_plus_structured.csv')
```

```python
test_df = pd.read_csv('data/CAD_test_halfsplit.csv', index_col=0)
test_df['HADM_ID'] = test_df['HADM_ID'].apply(lambda x: int(x))
test_df = test_df.merge(all_df_normalized, on='HADM_ID', how='left')
test_df['diabetesdx'] = test_df['diabetesdx'].fillna(0.0)
test_df['MED_ACE'] = test_df['MED_ACE'].fillna(0.0)
test_df['MED_INS'] = test_df['MED_INS'].fillna(0.0)
test_df['MED_BETA'] = test_df['MED_BETA'].fillna(0.0)
for col in all_df_normalized.columns:
    if col in ['HADM_ID']:
        continue
    if 'count' in col:
        test_df[col].fillna(0.0)
    test_df[col] = test_df[col].fillna(all_df_normalized[col].median())
test_df.to_csv('data/CAD_test_halfsplit_plus_structured.csv')
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import os
```

```python
train_df = pd.read_csv('data/CAD_train_halfsplit_leakedconcepts.csv')
val_df = pd.read_csv('data/CAD_val_halfsplit_leakedconcepts.csv')
test_df = pd.read_csv('data/CAD_test_halfsplit_leakedconcepts.csv')
```

```python
train_X = train_df[['HYPERTENSION','HYPERLIPIDEMIA','MEDICATION','DIABETES','OBESE']].replace(True,1).replace(False,0).fillna(0)
val_X = val_df[['HYPERTENSION','HYPERLIPIDEMIA','MEDICATION','DIABETES','OBESE']].replace(True,1).replace(False,0).fillna(0)
train_y = train_df['label']
val_y = val_df['label']
```

```python
clf = LogisticRegression(random_state=1).fit(train_X, train_y)
```

```python
from sklearn.metrics import matthews_corrcoef
print(clf.score(train_X, train_y))
print(clf.score(val_X, val_y))
print(matthews_corrcoef(clf.predict(val_X), val_y))
```
