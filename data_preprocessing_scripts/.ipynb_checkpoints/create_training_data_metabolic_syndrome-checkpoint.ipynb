```python
dirs_dct = dict(list(pd.read_csv('../directory_paths.csv')['paths'].apply(eval)))

admissions_df =  pd.read_csv(os.path.join(dirs_dct['mimic_dir'],'ADMISSIONS.csv'))

subjects_df = pd.read_csv(os.path.join(dirs_dct['mimic_dir'],'PATIENTS.csv'))

patients_df = pd.read_csv(os.path.join(dirs_dct['mimic_dir'],'PATIENTS.csv'))

notes_df =pd.read_csv(os.path.join(dirs_dct['mimic_dir'],'NOTEEVENTS.csv'))

#delete iserror
notes_df = notes_df[notes_df['ISERROR'].apply(lambda x: not x==1)]

admissions_df = admissions_df.merge(subjects_df[['SUBJECT_ID','DOB','GENDER']], on='SUBJECT_ID')

admissions_df['ADMITTIME'] = pd.to_datetime(admissions_df['ADMITTIME'])
admissions_df['DOB'] = pd.to_datetime(admissions_df['DOB'])
admissions_df['DEATHTIME'] = pd.to_datetime(admissions_df['DEATHTIME'])
admissions_df['DISCHTIME'] = pd.to_datetime(admissions_df['DISCHTIME'])

#Age in years
def get_age(row):
    try:
        return (row.ADMITTIME-row.DOB) / np.timedelta64(1, 's')/ 60./60/24/365
    except:
        return np.nan
admissions_df['AGE'] = admissions_df.apply(get_age, axis=1)

grouped = admissions_df.groupby('SUBJECT_ID')
#only look at patients with multiple admissions
only_multiple=grouped.filter(func=lambda x: len(x) >1, dropna=True)[['HADM_ID','SUBJECT_ID','DISCHTIME','ADMITTIME','AGE']]
#Find mean difference between admission and previous discharge
only_multiple.groupby('SUBJECT_ID').apply(lambda row:  row['ADMITTIME']-row['DISCHTIME'].shift(1)).mean()

only_multiple['interval_with_buffer']=only_multiple.apply(lambda row: pd.Interval(row['ADMITTIME'], (row['DISCHTIME']+pd.Timedelta('3 days'))) ,axis=1)

this_row=only_multiple.iloc[0]
new_hadm=[this_row['HADM_ID']]
for i in range(1,len(only_multiple)):
    prev_row=this_row
    this_row = only_multiple.iloc[i]
   
    if this_row['interval_with_buffer'].overlaps(prev_row['interval_with_buffer']):
        new_hadm.append(new_hadm[i-1])
    else:
        new_hadm.append(this_row['HADM_ID'])

only_multiple['new_HADM'] = new_hadm

only_multiple=only_multiple.merge(pd.DataFrame(only_multiple.groupby('new_HADM')['AGE'].apply(lambda group:group.min() )).reset_index(), how='left',on='new_HADM', suffixes=['','_new'])

admissions_df = admissions_df.merge(only_multiple[['HADM_ID','new_HADM','AGE_new']], how='left', on='HADM_ID')

admissions_df['new_HADM'] = admissions_df['new_HADM'].fillna(admissions_df['HADM_ID'])
admissions_df['AGE_new'] = admissions_df['AGE_new'].fillna(admissions_df['AGE'])

admissions_df[['HADM_ID','new_HADM']].to_csv(os.path.join(dirs_dct['data_dir'],'merged_HADMs.csv'),index=False)
```

```python
all_df['MED_ACE'] = all_df['MED_ACE'].fillna(False).apply(lambda x: 1.0 if x else 0.0)
all_df['MED_BETA'] = all_df['MED_BETA'].fillna(False).apply(lambda x: 1.0 if x else 0.0)
all_df['MED_INS'] = all_df['MED_INS'].fillna(False).apply(lambda x: 1.0 if x else 0.0)
all_df['MED_FIBR'] = all_df['MED_FIBR'].fillna(False).apply(lambda x: 1.0 if x else 0.0)
all_df['MED_STAT'] = all_df['MED_STAT'].fillna(False).apply(lambda x: 1.0 if x else 0.0)


all_df['diabetesdx'] = 0.0

all_df = all_df.merge(pd.DataFrame({'new_HADM':hyperlipidemia_adms, 'hyperlipidemia_dx':1}),on='new_HADM', how='left' )
all_df['hyperlipidemia_dx'] = all_df['hyperlipidemia_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':hypertension_adms, 'hypertension_dx':1}),on='new_HADM', how='left' )
all_df['hypertension_dx'] = all_df['hypertension_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':ins_adms, 'ins_dx':1}),on='new_HADM', how='left' )
all_df['ins_dx'] = all_df['ins_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':obesity_adms, 'obesity_dx':1}),on='new_HADM', how='left' )
all_df['obesity_dx'] = all_df['obesity_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':smoker_adms, 'smoker_dx':1}),on='new_HADM', how='left' )
all_df['smoker_dx'] = all_df['smoker_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':hyperglyceridemia_adms, 'hyperglyceridemia_dx':1}),on='new_HADM', how='left' )
all_df['hyperglyceridemia_dx'] = all_df['hyperglyceridemia_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':hyperglycemia_adms, 'hyperglycemia_dx':1}),on='new_HADM', how='left' )
all_df['hyperglycemia_dx'] = all_df['hyperglycemia_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':lowhdl_adms, 'lowhdl_dx':1}),on='new_HADM', how='left' )
all_df['lowhdl_dx'] = all_df['lowhdl_dx'].fillna(0)

all_df = all_df.merge(pd.DataFrame({'new_HADM':diabetes_adms, 'diabetes_dx':1}),on='new_HADM', how='left' )
all_df['diabetes_dx'] = all_df['diabetes_dx'].fillna(0)

rf_identifiers = {
    'new_HADM':lambda row: row['new_HADM'],
    'HYPERTENSION':lambda row: row['mean_sysbp']>=130 or row['mean_diasbp']>=85 or row['hypertension_dx']==1,
    'HYPERLIPIDEMIA': lambda row: row['max_ldl']>100 or row['max_cholesterol']>240 or row['hyperlipidemia_dx']>0, 
    'MED_ACE': lambda row: row['MED_ACE']==1,
    'MED_BETA': lambda row: row['MED_BETA']==1,
    'MED_INS': lambda row:  row['MED_INS']==1,
    'MED_FIBR': lambda row: row['MED_FIBR']==1,
    'MED_STAT': lambda row: row['MED_STAT']==1,
    'DIABETES': lambda row: row['diabetes_dx']==1,
    'HYPERGLYCERIDEMIA': lambda row: row['hyperglyceridemia_dx']==1 or row['MED_FIBR']==1 or row['mean_triglycerides']>150,
    'HYPERGLYCEMIA': lambda row: row['hyperglycemia_dx']==1,
    'LOWHDL': lambda row: (row['min_hdl']<40 and row['GENDER']=='M') or (row['min_hdl']<50 and row['GENDER']=='F') or  row['lowhdl_dx']==1,
    'OBESE': lambda row: row['max_weightkg']/(row['max_heightcm']/100)**2 > 30 or row['obesity_dx']>0,
    'SMOKER': lambda row: row['smoker_dx']>0,
    'MALE': lambda row: row['GENDER']=='M',
    'FEMALE': lambda row:  row['GENDER']=='F',
    'OLD': lambda row: row['AGE']>=65,
    'YOUNG': lambda row: row['AGE']<65,
    'BAD_CHOLESTEROL': lambda row: ( (row['min_hdl']<40 and row['GENDER']=='M') or (row['min_hdl']<50 and row['GENDER']=='F') or  row['lowhdl_dx']==1) or (
            row['max_ldl']>100 or row['max_cholesterol']>240 or row['hyperlipidemia_dx']>0 )
                            
}

negative_rf_identifiers = {
    'new_HADM':lambda row: row['new_HADM'],
    'HYPERTENSION':lambda row: row['mean_sysbp']<130 and row['mean_diasbp']<85 and row['hypertension_dx']==0,
    'HYPERLIPIDEMIA': lambda row: row['max_ldl']<100 and row['max_cholesterol']<240 and row['hyperlipidemia_dx']==0, 
    'MED_ACE': lambda row: row['MED_ACE']==0,
    'MED_BETA': lambda row: row['MED_BETA']==0,
    'MED_INS': lambda row:  row['MED_INS']==0,
    'MED_FIBR': lambda row: row['MED_FIBR']==0,
    'MED_STAT': lambda row: row['MED_STAT']==0,
    'DIABETES': lambda row: row['diabetes_dx']==0,
    'HYPERGLYCERIDEMIA': lambda row: row['hyperglyceridemia_dx']==0 and row['MED_FIBR']==0 and row['mean_triglycerides']<=150,
    'HYPERGLYCEMIA': lambda row: row['hyperglycemia_dx']==0,
    'LOWHDL': lambda row:( (row['min_hdl']>=40 and row['GENDER']=='M') or (row['min_hdl']>=50 and row['GENDER']=='F')) and row['lowhdl_dx']==0,
    'OBESE': lambda row: row['max_weightkg']/(row['max_heightcm']/100)**2 < 30 and row['obesity_dx']==0,
    'SMOKER': lambda row: row['smoker_dx']==0,
    'BAD_CHOLESTEROL': lambda row: (not (rf_identifiers['LOWHDL'](row) or rf_identifiers['HYPERLIPIDEMIA'](row))  ) #and ( row['MED_STAT']==0  )
    
}

rfs_df =  pd.DataFrame({rf:all_df.apply(fn,axis=1) for rf,fn in rf_identifiers.items()})
    

negative_rfs_df = pd.DataFrame({rf:all_df.apply(fn,axis=1) for rf,fn in negative_rf_identifiers.items()})

def has_metabolic_syndrome(row):
    criteria = [row['HYPERGLYCERIDEMIA'],   row['LOWHDL']  , row['HYPERTENSION'], row['DIABETES']]
    present_criteria = [x for x in criteria if x]
    return row['OBESE'] and len(present_criteria)>=2

def num_positive_criteria(row):
    criteria = [row['HYPERGLYCERIDEMIA'],   row['LOWHDL']  , row['HYPERTENSION'], row['DIABETES']]
    present_criteria = [x for x in criteria if x]
    return len(present_criteria)

def has_no_metabolic_syndrome(row):
    #call on negative_rfs_df
    criteria = [row['HYPERGLYCERIDEMIA'],   row['BAD_CHOLESTEROL']  , row['HYPERTENSION'], row['DIABETES']]
    present_criteria = [x for x in criteria if x]
    return row['OBESE'] or len(present_criteria)>=3

def num_negative_criteria(row):
    #call on negative_rfs_df
    criteria = [row['HYPERGLYCERIDEMIA'],   row['BAD_CHOLESTEROL']  , row['HYPERTENSION'], row['DIABETES']]
    present_criteria = [x for x in criteria if x]
    return len(present_criteria)

rfs_df['positive'] = rfs_df.apply(has_metabolic_syndrome, axis=1)
rfs_df['negative'] = negative_rfs_df.apply(has_no_metabolic_syndrome, axis=1)
```

```python
obese_patients = list(rfs_df[rfs_df['OBESE']]['SUBJECT_ID'].unique())
all_patients = set(admissions_df['SUBJECT_ID'].unique())
patients_ever_admitted_over_15 = set(admissions_df[admissions_df['AGE']>15]['SUBJECT_ID'].unique())
obese_patients_admitted_over_15 = set(obese_patients).intersection( set( patients_ever_admitted_over_15  )  )  
not_adult_patients = all_patients.difference(patients_ever_admitted_over_15)
patients_with_nonobese_bmi_admission = list(rfs_df[negative_rfs_df['OBESE']]['SUBJECT_ID'].unique())
nonobese_patients = list(set(patients_with_nonobese_bmi_admission).difference( set(obese_patients   ) ))
adult_nonobese_patients = set(nonobese_patients).intersection( set( patients_ever_admitted_over_15))
patients_with_two_risk_factors = list( rfs_df[rfs_df.apply(lambda row: num_positive_criteria(row)>1, axis=1)]['SUBJECT_ID'].unique() )
patients_with_three_negative_rfs = list( rfs_df[negative_rfs_df.apply(lambda row: num_negative_criteria(row)>2, axis=1)]['SUBJECT_ID'].unique() )
```

```python
def num_admissions(subject_ids):
    subject_ids_df=pd.DataFrame({'SUBJECT_ID':list(subject_ids)})
    return len(admissions_df.merge(subject_ids_df, on='SUBJECT_ID',how='inner')['new_HADM'].unique())
```

```python
#Calculate patients and admissions for those patients at each branch of flowchart
print('Age >15 at admission?')
print('YES: n= %d; h=%d' % (len(patients_ever_admitted_over_15),num_admissions(patients_ever_admitted_over_15) ) )
print('NO: n= %d; h=%d' % (len(not_adult_patients),num_admissions(not_adult_patients) ) )

print('Obese?')
print('YES: n= %d; h=%d' % (len(obese_patients_admitted_over_15),num_admissions(obese_patients_admitted_over_15) ) )
adults_not_known_obese = patients_ever_admitted_over_15.difference(obese_patients_admitted_over_15)
print('NO: n= %d; h=%d' % (len(adults_not_known_obese),num_admissions(adults_not_known_obese) ) )

print('evidence for 2 risk factors?')
two_rfs_and_obese_adult = obese_patients_admitted_over_15.intersection(patients_with_two_risk_factors)
print('YES: n= %d; h=%d' % (len(two_rfs_and_obese_adult),num_admissions(two_rfs_and_obese_adult) ) )
obese_adults_not_known_two_rfs = obese_patients_admitted_over_15.difference(two_rfs_and_obese_adult)
print('NO: n= %d; h=%d' % (len(obese_adults_not_known_two_rfs),num_admissions(obese_adults_not_known_two_rfs) ) )

print('negative evidence for 3 risk factors?')
not_two_positive_and_three_negative = obese_adults_not_known_two_rfs.intersection(patients_with_three_negative_rfs)
print('YES: n= %d; h=%d' % (len(not_two_positive_and_three_negative),num_admissions(not_two_positive_and_three_negative) ) )
no_three_rfs = obese_adults_not_known_two_rfs.difference(not_two_positive_and_three_negative)
print('NO: n= %d; h=%d' % (len(no_three_rfs),num_admissions(no_three_rfs) ) )

print('evidence for non-obese?')
print('YES: n= %d; h=%d' % (len(adult_nonobese_patients),num_admissions(adult_nonobese_patients) ) )
no_evidence_nonobese = adults_not_known_obese.difference(adult_nonobese_patients)
print('NO: n= %d; h=%d' % (len(no_evidence_nonobese),num_admissions(no_evidence_nonobese) ) )


positive_patients = two_rfs_and_obese_adult
negative_patients = not_two_positive_and_three_negative.union(adult_nonobese_patients)
excluded_patients = no_three_rfs.union(no_evidence_nonobese).union(not_adult_patients)
print('Positive: n= %d; h= %d' %(len(positive_patients),num_admissions(positive_patients)))
print('Negative: n= %d; h= %d' %(len(negative_patients),num_admissions(positive_patients)))
print('Excluded patients: n= %d; h= %d' %(len(excluded_patients),num_admissions(excluded_patients)))
all_patients=excluded_patients.union(negative_patients).union(positive_patients)
print(excluded_patients.intersection(negative_patients))
print(len(all_patients), len(admissions_df['SUBJECT_ID'].unique()))
```
