```python
#create data for joint oasis/mortality prediction/weekday prediction
merged['weekday']=pd.to_datetime(merged['CHARTDATE']).dt.day_name()
for day in merged['weekday'].unique():
    merged[day+'_score'] = merged['weekday'].apply(lambda x: 1 if x==day else 0)

merged
train_df = pd.read_csv('../data/mortality_train.csv')
train_df = merged[[col for col in merged.columns if 'score' in col or col in ['TEXT']]].merge(train_df, on='TEXT', how='inner')

train_df.to_csv('../data/mortality_and_oasis_train.csv')

#val
val_df = pd.read_csv('../data/mortality_val.csv')

val_df = merged[[col for col in merged.columns if 'score' in col or col in ['TEXT']]].merge(val_df, on='TEXT', how='inner')

val_df.to_csv('../data/mortality_and_oasis_val.csv')

#test
test_df = pd.read_csv('../data/mortality_test.csv')

test_df = merged[[col for col in merged.columns if 'score' in col or col in ['TEXT']]].merge(test_df, on='TEXT', how='inner')

test_df.to_csv('../data/mortality_and_oasis_test.csv')
```
